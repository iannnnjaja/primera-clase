```python
import numpy as np
#matrices de ejemplo

A = np.array([[5, 9, 3],
              [4, 7, 6],
              [2, 8, 9]])

B = np.array([[4, 8, 7],
              [6, 9, 3],
              [3, 2, 1]])

suma = A + B
print("Suma de matrices:\n", suma)
```

```python
producto_elemento = A * B   # producto elemento a elemento
print("Producto elemento a elemento:\n", producto_elemento)
```

```python
traza = np.trace(A)
print("Traza de A:", traza)
```

```python
transpuesta = A.T
print("Transpuesta de A:\n", transpuesta)
```

```python
maximo = np.max(A)
minimo = np.min(A)

print("Máximo:", maximo)
print("Mínimo:", minimo)
```

```python
suma_cuadrados = np.sum(A**2)
print("Suma del cuadrado de los elementos:", suma_cuadrados)

```

```python
A = np.array([[11,  7, -1],
              [-3,  5,  2],
              [ 4,  2, -2]])

B = np.array([[ 1,  1,  1],
              [10,  3, -5],
              [-6, -10, -3]])


#a)
resultado1 = 2*B + 3*A
print("2B + 3A =\n", resultado1)

#b)
C = B + 2*A
resultado2 = C @ C
print("(B + 2A)^2 =\n", resultado2)

#c)

resultado3 = (A @ A) + 3*B
print("A^2 + 3B =\n", resultado3)

```

```python
def trabajar_matriz(n, m):

    if n != m:
        return "La matriz no es cuadrada"

    if n < 3 or n > 5:
        return "Las dimensiones deben estar entre 3 y 5"

    T = np.random.randint(1, 101, size=(n, m))
    print("\nMatriz generada:\n", T)

    # Primera fila inversa
    print("\nPrimera fila invertida:", T[0][::-1])

    # Primera columna inversa
    print("Primera columna invertida:", T[:,0][::-1])

    # Primer y último elemento
    print("Primer elemento:", T[0,0])
    print("Último elemento:", T[-1,-1])

    # Máximo, mínimo, promedio diagonal y traza
    diagonal = np.diag(T)
    print("\nMáximo:", np.max(T))
    print("Mínimo:", np.min(T))
    print("Promedio diagonal:", np.mean(diagonal))
    print("Traza:", np.trace(T))

    # Desde T[1,1] hasta T[5,5]
    print("\nSubmatriz T[1,1] a T[n,n]:\n", T[1:n, 1:n])

    # Desde T[1,5] hasta T[5,1]
    if n >= 5:
        print("\nSubmatriz T[1,5] a T[5,1]:\n", T[0:5, 4:0:-1])

    # Suma de pares recorriendo filas
    suma_pares = np.sum(T[T % 2 == 0])
    print("\nSuma de elementos pares:", suma_pares)

    # Suma de impares recorriendo columnas
    suma_impares = np.sum(T[T % 2 != 0])
    print("Suma de elementos impares:", suma_impares)

    # Promedio filas y columnas
    print("\nPromedio de cada fila:", np.mean(T, axis=1))
    print("Promedio de cada columna:", np.mean(T, axis=0))

    # Antidiagonal
    anti = np.fliplr(T).diagonal()
    print("\nAntidiagonal:", anti)
    print("Promedio antidiagonal:", np.mean(anti))

trabajar_matriz(4,4)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 400)

C_positivos = [1, 2]
C_cero = [0]
C_negativos = [-1, -2]

plt.figure()

for C in C_positivos:
    y = (x - 1) + C * np.exp(-x)
    plt.plot(x, y, label=f"C = {C}")

for C in C_cero:
    y = (x - 1)
    plt.plot(x, y, label=f"C = {C}", linestyle='--')

for C in C_negativos:
    y = (x - 1) + C * np.exp(-x)
    plt.plot(x, y, label=f"C = {C}")

plt.title("Soluciones de la ecuación diferencial")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 2000)

def f(x):
    return (3*x + 7)/(x**2 - 4)

y = np.where(np.abs(x**2 - 4) < 0.1, np.nan, f(x))

plt.figure()
plt.plot(x, y, label="f(x) = (3x+7)/(x² - 4)")

plt.axvline(2, linestyle='--', label="x = 2")
plt.axvline(-2, linestyle='--', label="x = -2")
plt.axhline(0, linestyle='--', label="y = 0")

plt.ylim(-20, 20)
plt.legend()
plt.grid()

plt.show()
```
