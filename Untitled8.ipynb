```python
#ejercicio 1
def acceso_boveda(edad_juan, edad_pedro, num_nietos):
    # Condiciones del testamento
    condicion_a = (edad_juan > 35) or (edad_pedro < 50)
    condicion_b = ( (edad_juan + edad_pedro) % 2 == 0 )
    condicion_c = (num_nietos >= 1)

    # Si se cumplen todas las condiciones
    if condicion_a and condicion_b and condicion_c:
        return "SI"
    else:
        return "NO"
# Ejemplo de uso:
juan = int(input("Ingrese la edad de Juan: "))
pedro = int(input("Ingrese la edad de Pedro: "))
nietos = int(input("Ingrese el número de nietos de Manuel: "))


print("Acceso a la bóveda:", acceso_boveda(juan, pedro, nietos))
```

```python
#ejercicio 3
def es_primo(n):
    # Si el número es menor o igual a 1, se considera error
    if n <= 1:
        print("Error en el Número")
        return None
    # Verificación de número primo
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            print("False")
            return False
    print("True")
    return True
while True:
    try:
        numero = int(input("Ingrese un número entero: "))
        if es_primo(numero) is not None:
            break
    except ValueError:
        print("Debe ingresar un número entero válido.")
```

```python
# ejercicio 4

def procesar_numero(N):
    if N <= 100:
        print("Error en el número")
        return None

    ultimo = N % 10
    if ultimo in [4, 7, 9]:
        if N % 3 == 0:
            print("Es divisible por 3")
        else:
            print("No es divisible por 3")

    elif ultimo in [1, 5]:
        resultado = N // 10
        print(f"Resultado: {resultado}")

    else:
        residuo = N % 3
        print(f"Residuo: {residuo}")
while True:
    try:
        N = int(input("Ingrese un número entero mayor que 100: "))
        if procesar_numero(N):
            break
    except ValueError:
        print("Debe ingresar un número entero válido.")
```

```python
# ejercicio 5

def mcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

def mcm(a, b):
    return abs(a * b) // mcd(a, b)

a = int(input("Ingrese el primer número: "))
b = int(input("Ingrese el segundo número: "))

resultado_mcd = mcd(a, b)
resultado_mcm = mcm(a, b)

resultados = {
    "Número A": a,
    "Número B": b,
    "MCD": resultado_mcd,
    "MCM": resultado_mcm}

print("\n--- Resultados ---")
for clave, valor in resultados.items():
    print(f"{clave}: {valor}")
```

```python
# ejercicio 6

def filtrar_datos(lista):
    enteros = []
    cadenas = []
    booleanos = []
    flotantes = []

    for elemento in lista:
        if isinstance(elemento, bool):
            booleanos.append(elemento)
        elif isinstance(elemento, int):
            enteros.append(elemento)
        elif isinstance(elemento, float):
            flotantes.append(elemento)
        elif isinstance(elemento, str):
            cadenas.append(elemento)

    return enteros, cadenas, booleanos, flotantes

# Lista de datos
datos = [25, "Matemáticas", 3.141592, True, 8, False, "Nombres", 2.718, "UNAM", 0,
         9.81, "CdMx", 7e12, 12.5, "2025", True, "población", 0.01, 42, False]
enteros, cadenas, booleanos, flotantes = filtrar_datos(datos)


print("Enteros:", enteros)
print("Cadenas:", cadenas)
print("Booleanos:", booleanos)
print("Flotantes:", flotantes)
```

```python
# Ejercicio 7

import math

def calcular_distancia(lat1, lon1, lat2, lon2):
    """Calcula la distancia entre dos puntos en la Tierra usando la fórmula de Haversine.
    Entradas:
        lat1, lon1, lat2, lon2 -> coordenadas en grados decimales
    Salida:
        distancia en kilómetros (float)"""

    # Radio promedio de la Tierra en km
    R = 6378.1

    # Conversión de grados a radianes
    φ1, λ1, φ2, λ2 = map(math.radians, [lat1, lon1, lat2, lon2])

    # Diferencias de latitud y longitud
    delta_φ = φ2 - φ1
    delta_λ = λ2 - λ1

    # Definición de la función haversin
    haversin = lambda θ: math.sin(θ / 2) ** 2

    # Aplicación de la fórmula de Haversine
    d = 2 * R * math.asin(
        math.sqrt(
            haversin(delta_φ) +
            math.cos(φ1) * math.cos(φ2) * haversin(delta_λ)))

    return d

# PRUEBAS

#París - Roma
dist1 = calcular_distancia(48.9, 2.4, 41.9, 12.5)
print(f"París - Roma: {dist1:.0f} km")

# Ciudad de México - Nueva York
dist2 = calcular_distancia(19.43, -99.13, 40.71, -74.00)
print(f"Ciudad de México - Nueva York: {dist2:.0f} km")

#Ciudad de México - Buenos Aires
dist3 = calcular_distancia(19.43, -99.13, -34.61, -58.38)
print(f"Ciudad de México - Buenos Aires: {dist3:.0f} km")

```

```python
# ejercicio 8
def fibonacci(n):
    if n <= 0:
        print("Error en el número. Debe ser mayor que 0.")
        return None
    elif n == 1:
        print([0])
        return [0]

    elif n == 2:
        print([0, 1])
        return [0, 1]
    serie = [0, 1]
    for i in range(2, n):
        siguiente = serie[i - 1] + serie[i - 2]
        serie.append(siguiente)

    print("Serie de Fibonacci hasta", n, "términos:")
    print(serie)
    return serie

while True:
    try:
        n = int(input("Ingrese la cantidad de términos de la serie de Fibonacci: "))
        if fibonacci(n) is not None:
            break
    except ValueError:
        print("Debe ingresar un número entero válido.")
```

```python
#ejercicio 9
#listas
Fechas = ["2025-03-01","2025-03-02","2025-03-03","2025-03-04","2025-03-05",
          "2025-03-06","2025-03-07","2025-03-08","2025-03-09","2025-03-10"]

Zonas = ["guerrero","oaxaca","chiapas","michoacán","guerrero",
          "oaxaca","chiapas","guerrero","oaxaca","chiapas"]

magnitudes = ["3.8","4.2","5.0","2.9","4.8","3.5","4.1","5.3","4.0","3.0"]

prof_km = [15, 80, 12, 5, 35, 120, 60, 22, 10, 55]

magnitudes = list(map(float, magnitudes))

Zonas = list(map(lambda z: z.capitalize(), Zonas))
intensidad = []
for m in magnitudes:
    if m < 3.5:
        intensidad.append("Muy Leve")
    elif 3.5 <= m < 4.5:
        intensidad.append("Leve")
    elif 4.5 <= m < 5.5:
        intensidad.append("Moderado")
    else:
        intensidad.append("Fuerte")

conteo_por_zona = {}
for zona in Zonas:
    if zona in conteo_por_zona:
        conteo_por_zona[zona] += 1
    else:
        conteo_por_zona[zona] = 1

sismos = list(zip(Fechas, Zonas, magnitudes, prof_km))

sismos_ordenados = sorted(sismos, key=lambda x: x[2], reverse=True)

print("\n--- Tres sismos más grandes ---")
for fecha, zona, mag, prof in sismos_ordenados[:3]:
    print(f"{fecha} | {zona} | {mag:.1f} | {prof} km")

print("\n--- Conteo de eventos por zona ---")
for zona, cantidad in conteo_por_zona.items():
    print(f"{zona}: {cantidad}")

# Mostrar la lista de intensidades (opcional)
print("\n--- Intensidad de cada evento ---")
for i in range(len(Fechas)):
    print(f"{Fechas[i]} | {Zonas[i]} | {magnitudes[i]} | {intensidad[i]}")
```

```python
# ejercico 10
a = [2, -1, 4]
b = [4, -2, 8]
c = [-5, 2, 1]

def producto_escalar(v1, v2):
    return sum(x * y for x, y in zip(v1, v2))

def producto_vectorial(v1, v2):

    x = v1[1]*v2[2] - v1[2]*v2[1]
    y = v1[2]*v2[0] - v1[0]*v2[2]
    z = v1[0]*v2[1] - v1[1]*v2[0]
    return [x, y, z]

def relacion_vectores(v1, v2):
    escalar = producto_escalar(v1, v2)
    if escalar == 0:
        return "Perpendiculares"
    try:
        razones = [v1[i]/v2[i] for i in range(len(v1)) if v2[i] != 0]
        if all(r == razones[0] for r in razones):
            return "Paralelos"
    except ZeroDivisionError:
        pass
    return "Ninguna de las dos"



# Productos escalares
ab = producto_escalar(a, b)
bc = producto_escalar(b, c)
ca = producto_escalar(c, a)

# Productos vectoriales
axb = producto_vectorial(a, b)
bxc = producto_vectorial(b, c)
axc = producto_vectorial(a, c)

print("Producto Escalar")
print(f"ab = {ab}")
print(f"bc = {bc}")
print(f"ca = {ca}")

print("\n Producto Vectorial")
print(f"a x b = {axb}")
print(f"b x c = {bxc}")
print(f"a x c = {axc}")

print("\n Relaciones entre vectores")
print(f"a y b → {relacion_vectores(a, b)}")
print(f"b y c → {relacion_vectores(b, c)}")
print(f"a y c → {relacion_vectores(a, c)}")
```
